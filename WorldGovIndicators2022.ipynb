```python
import pandas as pd
import requests
from io import BytesIO

# URL of the file on GitHub
url = "https://github.com/paulo-marquesmorgado/EConcepts/raw/main/wgidataset.xlsx"

# Download the file content
response = requests.get(url)

# Read the content into a pandas DataFrame
with pd.ExcelFile(BytesIO(response.content)) as xls:
    # Load the "Data" sheet into df_wgi
    df_wgi = pd.read_excel(xls, sheet_name="Data")
    # Load the "Country" sheet into df_countries
    df_countries = pd.read_excel(xls, sheet_name="Country")

# Display the DataFrames
print("Dataframe from GitHub - Data:")
# print(df_wgi.head())
print("\nDataframe from GitHub - Country:")
# print(df_countries.head())
```

```python
# Assuming df_wgi is your existing DataFrame containing all columns
# Selecting only the desired columns
df_mywgi = df_wgi[['Country/Territory', 'Code', 'Dimention', '2022_Rank']].copy()

# Display the first few rows of df_mywgi
print(df_mywgi.head())

# Merge df_mywgi with df_countries on the 'Code' column
df_merged = df_mywgi.merge(df_countries[['CODE', 'Continent', 'Region 1']],
                           how='left', left_on='Code', right_on='CODE')

# Drop the redundant 'CODE' column
df_merged.drop(columns=['CODE'], inplace=True)

# Display the information about the merged DataFrame
print(df_merged)
```

```python
# Get unique values of the 'Dimention' column
unique_dimensions = df_merged['Dimention'].unique()

# Display the unique values
print(unique_dimensions)
```

```python
import pandas as pd

# Define the weights for each dimension
dimension_weights = {
    'Voice and Accountability': 1/6,
    'Political Stability and Absence of Violence/Terrorism': 1/6,
    'Government Effectiveness': 1/6,
    'Regulatory Quality': 1/6,
    'Rule of Law': 1/6,
    'Control of Corruption': 1/6
}

# Initialize an empty DataFrame to store the results
all_results = pd.DataFrame(columns=['Country/Territory', 'Region 1', 'Normalized Weighted Sum'])

# Loop through each unique combination of Continent and Region 1
for continent in df_merged['Continent'].unique():
    for region1 in df_merged[df_merged['Continent'] == continent]['Region 1'].unique():
        # Filter df_merged for the current combination of Continent and Region 1
        df_filtered = df_merged[(df_merged['Continent'] == continent) & (df_merged['Region 1'] == region1)]

        # Pivot the DataFrame to have Country/Territory per row and Dimension per column
        pivot_df = df_filtered.pivot(index='Country/Territory', columns='Dimention', values='2022_Rank')

        # Initialize a DataFrame to store the weighted sum
        weighted_sum = pd.DataFrame(index=pivot_df.index, columns=['Weighted Sum'])

        # Calculate the weighted sum for each country
        for dim, weight in dimension_weights.items():
            dim_weighted = pivot_df[dim] * weight
            weighted_sum['Weighted Sum'] = weighted_sum['Weighted Sum'].add(dim_weighted, fill_value=0)

        # Normalize the weighted sum by dividing by the sum of the weights
        sum_weights = sum(dimension_weights.values())
        weighted_sum_normalized = weighted_sum['Weighted Sum'] / sum_weights

        # Round the normalized weighted sum to 2 decimals using apply and lambda function
        weighted_sum_normalized = weighted_sum_normalized.apply(lambda x: round(x, 2))

        # Create a DataFrame to store the results for the current combination of Region 1
        result_df = pd.DataFrame({'Normalized Weighted Sum': weighted_sum_normalized})

        # Reset index to include 'Country/Territory' as a column
        result_df.reset_index(inplace=True)

        # Add 'Region 1' column with the current value
        result_df['Region 1'] = region1

        # Add 'Continent' column with the current value
        result_df['Continent'] = continent

        # Concatenate the results to the overall DataFrame
        all_results = pd.concat([all_results, result_df], ignore_index=True)

# Sort the DataFrame by Region 1 in ascending order and then by Normalized Weighted Sum in descending order
sorted_result_df = all_results.sort_values(by=['Region 1', 'Normalized Weighted Sum'], ascending=[True, False])

# Display the sorted DataFrame with the 'Normalized Weighted Sum'
# print(sorted_result_df[['Country/Territory', 'Region 1', 'Normalized Weighted Sum']])

# Group the DataFrame by 'Region 1'
grouped_df = sorted_result_df.groupby('Region 1')

# Iterate over each group and print the records followed by an empty line
for name, group in grouped_df:
    print(group.to_string(index=False))
    print()  # Print an empty line after each group
```
